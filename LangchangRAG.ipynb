```python
import os
from docx import Document as DocxDocument
import fitz  # PyMuPDF for PDF handling
from langchain.docstore.document import Document
from langchain.document_loaders import DirectoryLoader, TextLoader
```

```python
# Function to load a .docx document
def load_docx(file_path):
    doc = DocxDocument(file_path)
    text = ""
    for para in doc.paragraphs:
        text += para.text + "\n"
    return Document(page_content=text, metadata={"source": file_path})

# Function to load a PDF document
def load_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return Document(page_content=text, metadata={"source": file_path})
```

```python
# Specify the directory where your documents are stored
directory_path = "/content/drive/MyDrive/LangChain/Documents"
```

```python
# Debugging: Check if the directory exists
if not os.path.exists(directory_path):
    print(f"Directory does not exist: {directory_path}")
else:
    print(f"Directory exists: {directory_path}")

# Debugging: List files in the directory
print("\nFiles in directory:")
print(os.listdir(directory_path))
```

```python
# Manually load documents using custom loaders
def load_documents(directory_path):
    docx_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if f.endswith(".docx")]
    pdf_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if f.endswith(".pdf")]

    docx_documents = [load_docx(docx_file) for docx_file in docx_files]
    pdf_documents = [load_pdf(pdf_file) for pdf_file in pdf_files]

    # Use DirectoryLoader for .txt files
    text_loader = DirectoryLoader(directory_path, glob="*.txt", loader_cls=TextLoader)
    txt_documents = text_loader.load()

    # Combine all documents
    all_documents = docx_documents + pdf_documents + txt_documents
    return all_documents

# Load documents
documents = load_documents(directory_path)
```

```python
# Check if documents are loaded successfully
if documents:
    print(f"Successfully loaded {len(documents)} documents.")
    # Print a sample of the first document
    for n in range(len(documents)):
      print("\nSample document text:")
      print(documents[n].page_content[:500])  # Print first 500 characters of the document
      print("\nDocument metadata:")
      print(documents[n].metadata)
else:
    print("No documents were loaded.")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
# Initialize the RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,  # Maximum number of characters per chunk
    chunk_overlap=20 # Overlap between chunks
)

# Split the documents
all_chunks = []
for doc in documents:
    chunks = splitter.split_text(doc.page_content)
    for chunk in chunks:
        all_chunks.append(Document(page_content=chunk, metadata=doc.metadata))

# Print out some details about the chunks
print(f"Total chunks created: {len(all_chunks)}")
for i, chunk in enumerate(all_chunks[:5]):  # Print details for the first 5 chunks
    print(f"Chunk {i+1}:")
    print(f"Content: {chunk.page_content[:50]}...")  # Print first 200 characters of each chunk
    print(f"Metadata: {chunk.metadata}\n")
```

```python
from sentence_transformers import SentenceTransformer
# Initialize a local sentence transformer model for embeddings
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Function to generate embeddings for each chunk
def get_embedding(text):
    return embedding_model.encode(text)
```

```python
# Generate embeddings for each chunk and print a summary
embeddings = []
for i, chunk in enumerate(all_chunks):
    embedding = get_embedding(chunk.page_content)
    embeddings.append(embedding)

    # Print a summary of the embedding
    print(f"Embedding for Chunk {i+1}:")
    print(f"First 10 values: {embedding[:10]}")  # Print only the first 10 values for brevity
    print(f"Embedding Shape: {embedding.shape}")
    print()  # Print a blank line for readability

```

```python
# Optionally, we can print embeddings for only the first few chunks
print("Summary of Embeddings for First Few Chunks:")
for i in range(min(2, len(embeddings))):  # Print only the first 2 embeddings, adjust as needed
    print(f"Embedding for Chunk {i+1}:")
    print(f"First 10 values: {embeddings[i][:10]}")
    print(f"Shape: {embeddings[i].shape}")
    print()
```

```python
# Initialize Chroma client
client = chromadb.Client()
```

```python
# Ensure you have the correct dimension
embedding_dim = len(embeddings[0]) if embeddings else 0

# Check if collection exists and its dimension
existing_collection = client.get_collection(name="my_documents")

if existing_collection:
    current_dim = existing_collection.get("dimension", None)
    if current_dim != embedding_dim:
        # Handle dimension mismatch: delete and recreate collection or adjust
        client.delete_collection(name="my_documents")
        collection = client.create_collection(name="my_documents")
else:
    # Create collection if it does not exist
    collection = client.create_collection(name="my_documents")
```

```python
# Add documents to the ChromaDB collection
for i, chunk in enumerate(all_chunks):
    collection.add(
        ids=[str(i)],
        embeddings=[embeddings[i].tolist()],
        metadatas=[chunk.metadata],
        documents=[chunk.page_content]
    )
```

```python
# Implement a simple similarity search function using ChromaDB
def search(query, collection, top_n=3):
    query_embedding = get_embedding(query).tolist()
    results = collection.query(query_embeddings=[query_embedding], n_results=top_n)

    top_documents = results['documents']
    return top_documents
```

```python
# Example query
query = "what is name of greedy dog?"
top_documents = search(query, collection)

# Print the top results
for i, doc in enumerate(top_documents):
    print(f"Result {i+1}:")
    print(f"Content: {doc[:40]}...")  # Print first 200 characters of each chunk
    print()
```
